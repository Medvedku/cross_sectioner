## RSH

```python
import sqlite3
import pandas as pd
from pathlib import Path
import draw_functions as dfun

# 1. Setup Connection
db_path = Path.cwd().parent / 'scaper' / 'steel_engineering_final.db'
conn = sqlite3.connect(db_path)

# 2. Fetch one random row
query = "SELECT * FROM sections_rhs ORDER BY RANDOM() LIMIT 1"
random_profile = pd.read_sql(query, conn).iloc[0].to_dict()
conn.close()

# 3. Create the geometry
rhs_geom = dfun.create_rhs_geometry(
    h=random_profile['h'],
    b=random_profile['b'],
    t=random_profile['t'],
    r_out=random_profile['r_out'],
    r_in=random_profile['r_in'],
    section_id=random_profile['Section_ID']
)

# --- 4. Area Comparison ---
db_area = random_profile['A']
calc_area = rhs_geom.area
diff = abs(calc_area - db_area) / db_area * 100

print(f"Loaded Profile: {random_profile['Section_ID']}")
print(f"--- Area Comparison ---")
print(f"Database Area:  {db_area:>10.2f} mm²")
print(f"Shapely Area:   {calc_area:>10.2f} mm²")
print(f"Difference:     {diff:>10.4f} %")

# Trigger preview
rhs_geom
```

```python
import sqlite3
import pandas as pd
from pathlib import Path

# 1. Setup Connection
db_path = Path.cwd().parent / 'scaper' / 'steel_engineering_final.db'
conn = sqlite3.connect(db_path)

# 2. Fetch one random row from sections_ipe
query_val = "SELECT * FROM sections_ipeaa ORDER BY RANDOM() LIMIT 1"
random_profile = pd.read_sql(query_val, conn).iloc[0].to_dict()

# 3. Fetch Metadata for IPE profiles
query_meta = "SELECT column_name, description, unit FROM data_dictionary WHERE table_name = 'sections_ipe'"
df_meta = pd.read_sql(query_meta, conn)

conn.close()

# 4. Format for inspection
report = pd.DataFrame(random_profile.items(), columns=['column_name', 'Value'])
report = pd.merge(report, df_meta, on='column_name', how='left')
report['Result'] = report['Value'].astype(
    str) + " " + report['unit'].fillna("")

print(f"--- 📐 RANDOM IPE PROFILE INSPECTION: {random_profile['Section_ID']} ---")
print(report[['column_name', 'description', 'Result']].to_string(index=False))
```
