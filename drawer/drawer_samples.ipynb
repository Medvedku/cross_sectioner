## H & I with parallel flanges

```python
%load_ext autoreload
%autoreload 2

import sqlite3
import pandas as pd
from pathlib import Path
import draw_functions as dfun

def analyze_section(table_name):
    """
    Fetches a random profile from the given table, creates geometry,
    and compares the database area with the calculated Shapely area.
    """
    # 1. Fetch random profile
    db_path = Path.cwd().parent / 'scaper' / 'steel_engineering_final.db'
    conn = sqlite3.connect(db_path)
    
    # Using f-string for table name (safe here as input is internal/controlled)
    query = f"SELECT * FROM {table_name} ORDER BY RANDOM() LIMIT 1"
    random_profile = pd.read_sql(query, conn).iloc[0].to_dict()
    conn.close()

    # 2. Create geometry
    geom = dfun.create_i_section_geometry(
        h=float(random_profile['h']),
        b=float(random_profile['b']),
        tf=float(random_profile['tf']),
        tw=float(random_profile['tw']),
        r_root=float(random_profile['r_root'])
    )

    # 3. Compare Areas
    db_area = random_profile['A']
    calc_area = geom.area
    diff = abs(calc_area - db_area) / db_area * 100

    print(f"--- {table_name.upper()} ---")
    print(f"Loaded Profile: {random_profile['Section_ID']}")
    print(f"Database Area:  {db_area:>10.2f} mm²")
    print(f"Shapely Area:   {calc_area:>10.2f} mm²")
    print(f"Difference:     {diff:>10.4f} %")
    print("")

    return geom
```
